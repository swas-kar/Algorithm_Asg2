```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data from CSV
df = pd.read_csv('ObeservationNew.csv')
df
```

```python
# Plot data
plt.plot(df['Vertices'], df[' Avg.Time Taken DP'], label=' Avg.Time Taken DP', color='red')
plt.plot(df['Vertices'], df[' Avg.Time Taken Greedy'], label=' Avg.Time Taken Greedy',color='blue')

# Add labels and title
plt.xlabel('Vertices')
plt.ylabel('Time Taken')
plt.title('Comparison of Time Taken')

# Add legend
plt.legend()

# Show plot
plt.show()
```
